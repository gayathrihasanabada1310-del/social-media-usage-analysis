```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
data = {
    'Age': [18,22,25,30,35,40,28,19,23,31,27,21,45,50,33],
    'Platform': ['Instagram','Facebook','Twitter','Instagram','YouTube','Facebook',
                 'Instagram','Twitter','YouTube','Instagram','Facebook','Instagram',
                 'YouTube','Facebook','Twitter'],
    'Time_Spent': [120,90,60,150,200,80,110,70,180,140,95,130,210,75,65]
}

df = pd.DataFrame(data)
df.head()
```

```python
print(df.info())
print(df.describe())
```

```python
df.drop_duplicates(inplace=True)
print(df.isnull().sum())
```

```python
bins = [10, 20, 30, 40, 60]
labels = ['Teen', 'Young Adult', 'Adult', 'Senior']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
df.head()
```

```python
plt.figure()
sns.barplot(x='Platform', y='Time_Spent', data=df)
plt.title("Time Spent per Platform")
plt.xlabel("Platform")
plt.ylabel("Time (minutes)")
plt.show()
```

```python
plt.figure()
sns.countplot(x='Age_Group', hue='Platform', data=df)
plt.title("Age Group vs Platform Usage")
plt.show()
```

```python
plt.figure()
df['Platform'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Popular Platforms")
plt.ylabel("")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Platform', y='Time_Spent', data=df)
plt.title("Usage Distribution")
plt.show()
```
